```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
```

```python
boston=load_boston()
```

```python
print(boston.DESCR)
```

```python
df=pd.DataFrame(boston.data,columns=boston.feature_names)
df
```

```python
df.isnull()
```

```python
target=pd.DataFrame(boston.target,columns=['target'])
target
```

```python
df.describe().round(decimals=2)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df,target)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
test=boston.data[-30:]
```

```python
test
```

```python
predicted=reg.predict(test)
```

```python
predicted
```

```python
test1=boston.target[-30:]
```

```python
test1
```

```python
print("mean squared error is:",mean_squared_error(test1,predicted))
```
